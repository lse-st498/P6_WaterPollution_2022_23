```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import folium
import geopandas as gpd
```

```python
import geopy
```

```python

from global_land_mask import globe
```

```python
df = pd.read_csv('site.csv')
df
```

```python

mean_lat = df["lat"].mean()
mean_long = df["long"].mean()
map = folium.Map(location=[mean_lat, mean_long], zoom_start=6)

# Add markers for each site
for i, row in df.iterrows():
    folium.Marker(
        location=[row["lat"], row["long"]],
        popup=row["label"],
    ).add_to(map)


map
```

```python
ls = []
for i, row in df.iterrows():
    ls.append(globe.is_land(row["lat"], row["long"]))
```

```python
df['on_land'] = ls
```

```python
land_labels = df[df['on_land'] == True]['label'].tolist()
print(land_labels)
```

```python
df['pollutionRiskForecasting'] = df['pollutionRiskForecasting'].map({True: 1, False: 0})
df['pollutionRiskForecasting'].mean()
```

```python
df.groupby('district')['pollutionRiskForecasting'].mean()
```

```python
df.groupby('district')['pollutionRiskForecasting'].mean().sort_values().plot(kind='bar', figsize=(20,10))

plt.xlabel("District")
plt.ylabel("pollutionRiskForecasting")

plt.show()
```
