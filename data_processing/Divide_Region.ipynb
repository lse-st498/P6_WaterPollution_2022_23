```python
df_test_region = pd.read_csv('data/df_test_with_region.csv')
df_test_region
```

```python
x = []
y = []

for i in ['South West', 'South East', 'East of England', 'North East', 'North West', 
          'Yorkshire and The Humber', 'East Midlands']:
    x.append(i)
    
    # Filter data for the specific region
    region_data = df_test_region[df_test_region['region'] == i]
    
    # Calculate the total and positive counts from the filtered data
    total = len(region_data)
    positive = len(region_data[region_data['riskLevelLabel'] == 1])
    
    y.append(positive/total if total != 0 else 0)

print(x)
print(y)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y, marker='o', color='royalblue', linestyle='-')

# Titles and labels
plt.title('Proportion of positive label by Region')
plt.xlabel('Region')
plt.ylabel('Proportion of positive label')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('results/region_positive.png')
plt.show()
```
