```python
import pandas as pd
```

```python
site = pd.read_csv('data/site.csv')
site = site.iloc[:, 2:]
site
```

```python
DtoR = pd.read_csv('data/District_to_Region.csv')
DtoR.rename(columns={'LAD22NM': 'district'}, inplace=True)
DtoR.rename(columns={'RGN22NM': 'region'}, inplace=True)
DtoR
```

```python
site_with_region = pd.merge(site, DtoR[['district', 'region']], on='district', how='left')
site_with_region
```

```python
site_with_region['region'].value_counts()
```

```python
site_with_region.to_csv('data/site_with_region.csv')
```
